```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')
```

Dataset Source --- https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/

```python
df = pd.read_csv("agaricus-lepiota.csv",header=None)   #opened in notepad the data file and saved as csv
```

```python
#From the documentation it is clear that the target column is column number 0 where e = edible and p = poisonous
```

```python
# names taken from Documentation
names = ['class','cap-shape','cap-surface','cap-color','bruises','odor','gill-attachment','gill-spacing','gill-size','gill-color',
         'stalk-shape','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring',
         'stalk-color-below-ring','veil-type','veil-color','ring-number','ring-type','spore-print-color','population','habitat']
```

```python
len(names)
```

```python
df.columns = names
```

```python
df
```

### Data Preprocessing and Model Building 

```python
lbl=LabelEncoder()
```

```python
for col in df.columns:
    df[col] = lbl.fit_transform(df[col])
```

```python
df.head()
```

```python
#separating the attibutes and class
x = df.iloc[:,1:23]
y = df['class']
```

```python
y.head()
```

```python
from collections import Counter
# The output matches with the documentation provided in UCL repository where Class Distribution: 
#                                             --    edible: 4208 (51.8%)
#                                             -- poisonous: 3916 (48.2%)
#                                             --     total: 8124 instances
Counter(y) 
```

```python
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state = 47)   #test train split
```

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

### Kmeans Clustering 

```python
k_means = KMeans(n_clusters=2)
```

```python
k_means.fit(x_train, y_train)
```

```python
k_means_pred = k_means.predict(x_test)
```

```python
print("accuracy_score: ",accuracy_score(y_test,k_means_pred))
```

```python
print(confusion_matrix(y_test,k_means_pred))
```

```python
print(classification_report(y_test,k_means_pred))
```

### Logistic Regression Model 

```python
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
lg_pred = lg.predict(x_test)
```

```python
print("accuracy_score: ",accuracy_score(y_test,lg_pred))
```

```python
print(confusion_matrix(y_test,lg_pred))
```

```python
print(classification_report(y_test,lg_pred))
```

### Saving Best Model

```python
# Here best model is Logistic Regression
import joblib

# Saving the model as a pickle in a file
joblib.dump(lg,"mushroom_dataset_classification.pkl")
```
